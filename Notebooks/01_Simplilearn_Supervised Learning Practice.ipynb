```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Data
df = pd.read_csv('/content/titanic_dataset.csv')
```

```python
# Check First 5 Rows of Dataset
df.head()
```

```python
# to check statitical description of numerical features
df.describe()
```

```python
# to check complete information of dataset
df.info()
```

```python
# Create a new feature for people who are traveling alone
df['TravelAlone']=np.where((df["SibSp"]+df["Parch"])>0, 0, 1)
```

```python
df.head()
```

```python
# Delete the variables that do not contribute in our data
df1 = df.drop(['PassengerId','Name','Ticket', 'SibSp', 'Cabin','Parch'], axis=1)
```

```python
# Check missing Values
df1.isnull().sum()
```

```python
# Fill missing values of age column
df1['Age'].fillna(df1['Age'].median(), inplace=True)
```

```python
# Fill missing values of embarked column
df1['Embarked'].fillna(df1['Embarked'].mode()[0], inplace=True)
```

```python
# Check missing Values
df1.isnull().sum()
```

```python
# Replace these columns with dummy values
df_titanic = pd.get_dummies(df1, columns=['Sex', 'Embarked', 'Pclass'], drop_first=True)
```

```python
df_titanic.head()
```

```python
# Splitting Data into X and Y variable
X = df_titanic.drop('Survived', axis=1)
y = df_titanic['Survived']
```
