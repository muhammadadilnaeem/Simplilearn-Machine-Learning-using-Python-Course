### **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import warnings
warnings.filterwarnings('ignore')
```

### **Make a Data**

```python
X, y = make_blobs(n_samples=100, centers=2, random_state=2, cluster_std=1.5)
```

- This code generates a synthetic dataset with 100 samples, divided into 2 clusters. The clusters are centered around two points, with a standard deviation of 1.5, making them relatively spread out. The `random_state=2` ensures reproducibility of the results.

### **Plot the Clusters**

```python
plt.scatter(X[:, 0], X[:, 1], c=y,s=50,cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters')
plt.show()
```

### **Load the Naive Bayse Model**

```python
model = GaussianNB()
```

### **Fit the Model**

```python
model.fit(X, y)
```

### **Generating New Data and Making Predictions with a Model**

```python
rrg = np.random.RandomState(0)
Xnew = [[-6, -14]] + [14, 18] * rrg.rand(2000, 2)
ynew = model.predict(Xnew)
```

- This code snippet generates a new dataset `Xnew` with 2,000 samples. It starts with the point `[-6, 14]` and adds random variations based on a uniform distribution scaled by `[14, 18]`. The `model.predict(Xnew)` line then uses a pre-trained model to make predictions for these new data points, storing the results in `ynew`.

### **Visualizing Clusters with Original and New Data Points**

```python
plt.scatter(X[:, 0], X[:, 1], c=y,s=50,cmap='rainbow')
llm=plt.axis()

plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20,cmap='rainbow', alpha=0.1)  # Use ynew for colors
plt.axis(llm)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters')
plt.show()
```

### **Retrieving and Rounding Predicted Probabilities for New Data Points**

```python
yproba = model.predict_proba(Xnew)
yproba[-8:].round(2)
```

- This code retrieves the predicted probabilities for the last 8 samples in `Xnew` using the `predict_proba` method of the model. The results are rounded to two decimal places for clarity.
