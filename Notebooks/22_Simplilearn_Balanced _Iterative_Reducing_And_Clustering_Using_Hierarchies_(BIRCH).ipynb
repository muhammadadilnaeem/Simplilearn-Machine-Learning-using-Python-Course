**What is BIRCH in machine learning?**
- BIRCH (balanced iterative reducing and clustering using hierarchies) is an unsupervised data mining algorithm used to perform hierarchical clustering over particularly large data-sets.

### **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.cluster import Birch

import warnings
warnings.filterwarnings('ignore')
```

### **Make Practice Dataset**

- This code generates synthetic data using the make_blobs function. It creates 450 samples grouped into 6 clusters, each with a standard deviation of 0.7. The random_state=0 ensures reproducibility of the data. The result is two arrays: X for the data points and clusters for cluster labels.

```python
X,clusters = make_blobs(n_samples=450,centers=6,cluster_std=0.7,random_state=0)
```

### **Plot Clusters**

- This code creates a scatter plot of the synthetic data points `X`. Each point is plotted using its first and second features. The plot includes a title "Practice Dataset" and displays the points with a slight transparency (`alpha=0.7`) and black edges (`edgecolors='k'`).

```python
plt.scatter(X[:,0],X[:,1],alpha=0.7,edgecolors='k')
plt.title('Practice Dataset Before Applying BIRCH')
plt.show()
```

### **Let's Implement BIRCH**

```python
birch = Birch(branching_factor=50,n_clusters=None, threshold=1.5)
birch.fit(X)
```

```python
labels = birch.predict(X)
```

### **Let's Plot to See how Cluster are Formed**

```python
plt.scatter(X[:,0],X[:,1],c=labels,alpha=0.7,cmap='rainbow',edgecolors='k')
plt.title('Practice Dataset After Applying BIRCH')
plt.show()
```
