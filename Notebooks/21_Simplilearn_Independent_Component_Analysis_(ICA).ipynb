### **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import FastICA

import warnings
warnings.filterwarnings('ignore')
```

### **Load the Dataset**

```python
df = pd.read_csv('/content/employe_survey_data.csv')
```

### **First 5 Rows of Dataset**

### **Shape of the Dataset**

```python
print(f"This dataset consist of {df.shape[0]} rows and {df.shape[1]} columns.")
```

## **Implement ICA With 2 Components**

```python
ICA = FastICA(n_components=2)
IndependentComponents = ICA.fit_transform(df.values)
```

```python
# Print Finnal Independent components for that
print('## Final  Independent Components')
reduced_df = pd.DataFrame(IndependentComponents, columns=['Independent Component 1', 'Independent Component 2'])
reduced_df.head()
```
