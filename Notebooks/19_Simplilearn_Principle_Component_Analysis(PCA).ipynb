### **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')
```

### **Load the Dataset**

```python
b_cancer = load_breast_cancer()
print(b_cancer['DESCR'])
```

```python
df = pd.DataFrame(b_cancer['data'], columns=b_cancer['feature_names'])
df.head()
```

### **Split the Data into Train and Test**

```python
X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
```

### **Standardize the Dataset**

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

- This code standardizes the features of the training and test datasets `X_train` and `X_test`. It uses `StandardScaler` to fit the scaler on `X_train`, then transforms both `X_train` and `X_test` so that they have a mean of 0 and a standard deviation of 1.

### **Apply PCA to Training Dataset**

```python
pca = PCA()

X_train_pca = pca.fit_transform(X_train_std)
```

```python
exp_var_pca = pca.explained_variance_ratio_
```

```python
cum_sum_eigenvalues = np.cumsum(exp_var_pca)
```

### **Visualization of Explained Variance in PCA using Bar and Step Plots**

- This code snippet creates a bar plot to display the individual explained variance of principal components and a step plot for the cumulative explained variance, enhancing the understanding of how much variance each component captures in a Principal Component Analysis (PCA).

```python
plt.bar(range(0, len(exp_var_pca)), exp_var_pca, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0, len(cum_sum_eigenvalues)), cum_sum_eigenvalues, where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
