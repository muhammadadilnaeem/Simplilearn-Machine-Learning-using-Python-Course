### **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/position_salaries.csv')
```

### **Extracting Features and Labels from the DataFrame**

```python
X = df.iloc[:, 1:2].values
y = df.iloc[:, 2].values
```



1. **Feature Extraction**:
   - `X = df.iloc[:, 1:2].values`: Selects the second column (index 1) from `df`. The result is converted to a NumPy array and assigned to `X`. Note that using `1:2` returns a 2D array.

2. **Label Extraction**:
   - `y = df.iloc[:, 2].values`: Selects the third column (index 2) from `df` and converts it to a NumPy array, assigning it to `y`.

In summary, `X` contains the feature data, while `y` contains the corresponding labels for a machine learning task.

### **Split data into Train and Test**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### **Fit Linear Regression Model**

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

### **Visualize Linear Regression model Performance**

```python
def viz_linear():
    plt.scatter(X, y, color = 'red')
    plt.plot(X, lin_reg.predict(X), color = 'blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    return
```

- The function `viz_linear()` creates a scatter plot of data points (X, y) in red and overlays a linear regression line in blue. It also sets the title and axis labels before displaying the plot.

```python
viz_linear()
```

## **Fit Polynomial Regression**

- As Linear Regression is not performing well on data so now we will try Polynomial Regresion.

```python
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)
```

- This code snippet performs polynomial regression. It first transforms the feature set `X` into a polynomial feature set of degree 4 using `PolynomialFeatures`. Then, it creates a linear regression model and fits it to the transformed polynomial features (`X_poly`) and the target variable `y`.

### **Visualize Polynomial Regression model Performance**

```python
def viz_polynomial():
    plt.scatter(X, y, color = 'red')
    plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color = 'blue')
    plt.title('Polynomial Regression')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    return
```

```python
viz_polynomial()
```

### **Make Prediiction with Polynomial Regression**

```python
lin_reg.predict([[6.5]])
```

- The code `lin_reg.predict([[6.5]])` uses the linear regression model `lin_reg` to predict the target value (e.g., salary) for an input feature value of 6.5 (e.g., position level). The result will be the model's estimated output for that input.

```python
pol_reg.predict(poly_reg.fit_transform([[6.5]]))
```

- The code `pol_reg.predict(poly_reg.fit_transform([[6.5]]))` uses the polynomial regression model `pol_reg` to predict the target value for an input feature value of 6.5. It first transforms the input into polynomial features using `poly_reg.fit_transform`, then passes these features to the model to obtain the prediction.
